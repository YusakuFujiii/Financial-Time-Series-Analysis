## Simulating a random walk

Definition : 

$X_t = X_{t-1} + W_t,$ where $W_t \sim WN(0,\sigma^{2})$ and $X_0 = 0$

Expected value :

$E[X_t] = 0$

Variance :

$Var[X_t] = t\sigma^2$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# make a random walk time series 
white_n = np.random.normal(0,1,500)
random_w = np.cumsum(white_n)
num_as_date = np.arange(1,501)

# make a fig object
# facecolor of fig
fig, ax = plt.subplots(figsize = (12,5), facecolor = 'black')

# fig title 
fig.suptitle('Random walk', color ='cyan')

# facecolor of ax
ax.set_facecolor('black')

# label color 
ax.xaxis.label.set_color('c')
ax.yaxis.label.set_color('c')

# tick color 
ax.tick_params(axis='x', colors='c')
ax.tick_params(axis='y', colors='c')

# define a function to animate the time series
y = []
x = [] 
def animate(frame):
    y.append(random_w[frame])
    x.append(num_as_date[frame])
    plt.cla()
    ax.plot(x, y, color = 'c')
ani = FuncAnimation(fig, animate, range(500), interval = 50)
ani.save("rw.gif", writer="pillow")

```
