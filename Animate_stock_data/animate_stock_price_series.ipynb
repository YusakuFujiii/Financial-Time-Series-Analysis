## Animate Stock Price Series with Python & Matplotlib

```python
# import libraries 
import datetime
import pandas as pd


import numpy as np
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# set the range of data
start = datetime.date.today() - datetime.timedelta(days = 365)
end = datetime.date.today()

# fetch Google stock price from yahoo finance for one year 
df = pd.DataFrame()
df['goog'] = web.DataReader(['GOOG'], 'yahoo', start, end)['Adj Close']

# make a fig object
# facecolor of fig
fig, ax = plt.subplots(figsize = (12,5), facecolor = 'black')

# fig title 
fig.suptitle('Google stock price', color ='cyan')

# facecolor of ax
ax.set_facecolor('black')

# label color 
ax.xaxis.label.set_color('c')
ax.yaxis.label.set_color('c')

# tick color 
ax.tick_params(axis='x', colors='c')
ax.tick_params(axis='y', colors='c')

# animate the data
x = []
y = []

def update(frame):
    
    x.append(df.index[frame])
    y.append(df.iloc[frame,0])
    plt.cla()
    #ax.plot(x, y, color = 'cyan', label = 'Google Stock Price')
    ax.plot(x,y, color = 'cyan')
    # legend color 
    #plt.legend(labelcolor = 'linecolor', facecolor = 'black', edgecolor = 'black')

ani = FuncAnimation(fig, update, range(df.shape[0]), interval = 50)
ani.save("goog.gif", writer="pillow")
```
